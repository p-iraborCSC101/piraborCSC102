```python
#project 2

class Trapezium:
    def __init__(self, height, base1, base2):
        self.height = height
        self.base1 = base1
        self.base2 = base2
        
    def trapezium_area(self):
        area = (self.height / 2) * (self.base1 + self.base2)
        print(f'The area is {area}')
        
class Rhombus:
    def __init__(self, d1, d2):
        self.d1 = d1
        self.d2 = d2
        
    def rhombus_area(self):
        area = 0.5 * self.d1 * self.d2
        print(f'The area is {area}')
        
class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height
        
    def parallelogram_area(self):
        area = self.base * self.height
        print(f'The area is {area}')
        
class Cube:
    def __init__(self, length):
        self.length = length
        
    def cube_area(self):
        area = (self.length ** 2) * 6
        print(f'The area is {area}')
        
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        
    def cylinder_volume(self):
        volume = 3.14 * (self.radius ** 2) * self.height
        print(f'The volume is {volume}')
        
eqn = int(input('Enter want you want to calculate: \n1. Area of Trapezium \n2.Area of Rhombus \n3.Area of Parallelogram \n4.Area of Cube \n5.Volume of Cylinder \n'))

if eqn == 1:
    h = int(input('Enter height: '))
    b1 = int(input('Enter first base: '))
    b2 = int(input('Enter second base: '))
    
    trape = Trapezium(h, b1, b2)
    trape.trapezium_area()
    
elif eqn == 2:
    diag1 = int(input('Enter first diagonal: '))
    diag2 = int(input('Enter second diagonal: '))
    
    rhom = Rhombus(diag1, diag2)
    rhom.rhombus_area()
    
elif eqn == 3:
    base = int(input('Enter base: '))
    height = int(input('Enter height: '))
    
    parallel = Parallelogram(base, height)
    parallel.parallelogram_area()
    
elif eqn == 4:
    l = int(input('Enter length: '))
    cube = Cube(l)
    cube.cube_area()
    
else:
    r = int(input('Enter radius: '))
    h = int(input('Enter height: '))
    
    cylinder = Cylinder(r, h)
    cylinder.cylinder_volume()
    
```
